## Using Google API for snap point of dot grid to road and check for availability Street View

## Settings

- around_list = (0,60, 120, 180, 240, 300)
- size = '640x640'
- location = lat,lon
- fov = 100
- heading = around_list

```python
around_list = (0,60, 120, 180, 240, 300)
```

## Create dir

```python
try: 
    os.mkdir("data_one") 
    os.mkdir("data_two")
    print('done')
except OSError as error: 
    #print(error) 
    print('folder exists') 
    pass
```

```python
path_one = os.getcwd()+"/data_one/"
path_two = os.getcwd()+"/data_two/"
```

```python
def snap_check_download(coord, path):
    for y,x in coord:
        url_snap = 'https://roads.googleapis.com/v1/snapToRoads?path={},{}&interpolate=true&key={}'.format(y,x, apikey)
        response_snap = requests.get(url_snap)
        ans_snap = response_snap.json()
        time.sleep(5)
        if len(ans_snap) !=0:
            y_,x_ = ans_snap['snappedPoints'][0]['location']['latitude'],ans_snap['snappedPoints'][0]['location']['longitude']  
            url_check = 'https://maps.googleapis.com/maps/api/streetview/metadata?size=1x1&location={},{}&fov=0&heading=0&pitch=0&key={}'.format(y_,x_, apikey)
            response_check = requests.get(url_check)
            ans_check = response_check.json()
            if ans_check['status'] != 'ZERO_RESULTS':
                print('ok')
                for degr in around_list:
                    try:
                        url = "https://maps.googleapis.com/maps/api/streetview?size=240x240&location={},{}&fov=180&heading={}&pitch=0&key={}".format(y_,x_, degr, apikey)
                        #earli use640x640
                        response_gg = requests.get(url)
                        img = Image.open(BytesIO(response_gg.content))
                        img.save(path + "stview_{}_{}_{}.jpg".format(degr, y_, x_)) 
                        time.sleep(10) 
                    except:
                        time.sleep(5) 
                        response_gg = requests.get(url)
                        img = Image.open(BytesIO(response_gg.content))
                        img.save(path + "stview_{}_{}_{}.jpg".format(degr, y_, x_)) 
                        time.sleep(10)
            else:
                print('pass2')
                pass
        else:
            print('pass1')
```

```python
snap_check_download(ter_one_list, path_one)
```

```python
snap_check_download(ter_two_list, path_two)
```
